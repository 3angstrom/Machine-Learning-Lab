Calculate the mean value of each row of the data set and normalized the
value within the range 0 to 1, and the resultant value will be considered as

```python
max_dep = data['Dependent Feature'].max()
```

```python
min_dep = data['Dependent Feature'].min()
```

```python
data['Dependent Feature'] = data['Dependent Feature'].apply(lambda x: (x-min_dep)/(max_dep - min_dep))
```
