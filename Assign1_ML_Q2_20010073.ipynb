```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

# 1. Importing data CGU.csv'

```python
data = pd.read_csv('CGU.csv')
data
```

# Defining the target value

```python
data.median(axis = 1)
```

```python
data['Dependent Feature'] = data.median(axis=1)
```

```python
data
```

```python
#Using MIN-MAX Scaling
max_dep = data['Dependent Feature'].max()
min_dep = data['Dependent Feature'].min()
data['Dependent Feature'] = data['Dependent Feature'].apply(lambda x: (x-min_dep)/(max_dep - min_dep))
data
```

```python
data['Dependent Feature'].max()
```

```python
data['Dependent Feature'].min()
```

```python
data.describe()
```

# Correlation check

```python
#cleaning data
data.dropna(axis = 1, inplace = True, how='any')
```

```python
data.corr()
```

Cleaning and splitting of DATA

```python
x = data.iloc[: , :-1]

y = data.iloc[: , -1]

x.shape, y.shape
```

# Prepare a prediction model using multiple regression

```python
from sklearn.preprocessing import Normalizer 
norm = Normalizer() 
columns = x.columns 
x = norm.fit_transform(x) 
x = pd.DataFrame(x, columns = columns) 
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
```

# Model Testing

```python
predictions = model.predict(x_test)
model.score(x_test, y_test)
```

```python
print(predictions)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print('mean_squared_error : ', mean_squared_error(y_test, predictions))
print('mean_absolute_error : ', mean_absolute_error(y_test, predictions))
```

# Model Deletion Diagnostics (checking Outliers or Influencers)
Two Techniques : 1. Cook's Distance & 2. Leverage value

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
```

```python
#Model statistics
model = sm. OLS (y_train, sm.add_constant (x_train)). fit ()
print_model = model. summary ()
print (print_model)
```

```python
# 1. Cook's Distance: If Cook's distance > 1, then it's an outlier
# Get influencers using cook's distance
(c,_)=model.get_influence().cooks_distance
c
```

```python
# Plot the influencers using the stem plot
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(x_train)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
# Index and value of influencer where C>0.5
np.argmax(c) , np.max(c)
```

```python
influence = model.get_influence()
(cooks, p) = influence.cooks_distance
leverage = influence.hat_matrix_diag
from statsmodels.formula.api import ols
col = ['leverage']
leverage = pd.DataFrame (data = leverage , columns = col)
cooks = pd.DataFrame (data = cooks , columns = ['Cooks_dis'])
levcook = pd.concat ( [leverage, cooks],axis = 1)
Result = pd.concat ( [y, levcook] ,axis = 1)
Result=Result.rename (columns={'Dependent Feature': 'Dependent_Feature'})
Result
```

```python
#Identify high leverage
#point with leverage = (2k+2) /n
#k = number of predictors
#n= number of observations
k = x.shape [1]
n = x.shape [0]
outlier = pd.DataFrame ((Result.Dependent_Feature[abs(Result.leverage) > ((2*k)+2) /n])) 
outlier
```

```python
# Leverage Cuttoff Value = 3*(k+1)/n ; k = no.of features/columns & n = no. of datapoints
k=data.shape[1]
n=data.shape[0]
leverage_cutoff = (3*(k+1))/n
leverage_cutoff
```

```python
cutoff=(Result. loc[:, "Cooks_dis"].mean ()) *3
outlier1=pd.DataFrame((Result.Dependent_Feature [abs (Result.Cooks_dis) > cutoff])) 
print (outlier1)
```

```python
Influential = pd.merge(outlier1, outlier, left_index = True, right_index = True) 
print(Influential)
```

# Improving the model

```python
#Converting index series into list
index = Influential.index.tolist ()

# Dropping all the rows which have INFLUENTIAL's in it
data_copy = x.copy()
data_copy.drop(index = index, axis = 0)
data_copy.reset_index ()
```

```python
X_new_train, X_new_test, Y_new_train, Y_new_test =train_test_split (data_copy, y,test_size = 0.2, random_state = 0)
new_Lr = LinearRegression ()
new_Lr.fit (X_new_train, Y_new_train)
```

```python
predictions_new = new_Lr.predict(X_new_test)
```

```python
print('mean_squared_error : ', mean_squared_error(Y_new_test, predictions_new))
print('mean_absolute_error : ', mean_absolute_error(Y_new_test, predictions_new))
```
