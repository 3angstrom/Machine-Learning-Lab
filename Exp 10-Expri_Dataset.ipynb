Using SVM on Exp_Dataset with PCA

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv('Expri_Dataset.csv')
```

```python
data.info
```

```python
data.describe()
```

```python
from sklearn.preprocessing import Normalizer
```

```python
x = data.iloc[: , :-1]
y = data.iloc[:, -1]
```

```python
x.head()
```

```python
norm = Normalizer()
columns = x.columns
x_norm = norm.fit_transform(x)
x_norm = pd.DataFrame(x_norm, columns = columns)
x_norm.head()
```

```python
x_norm.corr().style.background_gradient(cmap = 'coolwarm')
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca.fit(x_norm)
x_pca = pca.transform(x_norm)
```

```python
x_pca = pd.DataFrame(x_pca, columns = ['Feature_1', 'Feature_2'])
x_pca.head()
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
y = label_encoder.fit_transform(y)
y = pd.DataFrame(y, columns = ['Target'])
```

```python
x_pca = pd.concat([x_pca, y], axis = 1)
x_pca
```

```python
sns.scatterplot(data = x_pca, x='Feature_1', y='Feature_2', hue = 'Target')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x_pca, y, test_size=0.3, random_state=0)
```

```python
from sklearn.svm import SVC
 
# Building a Support Vector Machine on train data
svc_model = SVC(C= .1, kernel='linear', gamma= 1, probability=True)
svc_model.fit(X_train, y_train)
 
prediction = svc_model .predict(X_test)
# check the accuracy on the training set
print(svc_model.score(X_train, y_train))
print(svc_model.score(X_test, y_test))
```

```python
#Building a Support Vector Machine on train data(changing the kernel)
svc_model_poly  = SVC(kernel='poly')
svc_model_poly.fit(X_train, y_train)
 
prediction_poly = svc_model.predict(X_test)
 
print(svc_model_poly.score(X_train, y_train))
print(svc_model_poly.score(X_test, y_test))
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
np.set_printoptions(precision=2)

# Plot non-normalized confusion matrix
titles_options = [
    ("Confusion matrix, without normalization", None),
    ("Normalized confusion matrix", "true"),
]
for title, normalize in titles_options:
    disp = ConfusionMatrixDisplay.from_estimator(
        svc_model,
        X_test,
        y_test,
        cmap=plt.cm.Blues,
        normalize=normalize,
    )
    disp.ax_.set_title(title)

    print(title)
    print(disp.confusion_matrix)

plt.show()
```

```python
pred_prob = svc_model.predict_proba(X_test)
from sklearn.metrics import roc_curve

# roc curve for models
fpr1, tpr1, thresh1 = roc_curve(y_test, pred_prob[:,1], pos_label=1)
from sklearn.metrics import roc_auc_score

# auc scores
auc_score1 = roc_auc_score(y_test, pred_prob[:,1])
auc_score1
```

```python
random_probs = [0 for i in range(len(y_test))]
p_fpr, p_tpr, _ = roc_curve(y_test, random_probs, pos_label=1)
```

```python
plt.style.use('seaborn')

# plot roc curves
plt.plot(fpr1, tpr1, linestyle='--',color='orange', label='SVM')
plt.plot(p_fpr, p_tpr, linestyle='--', color='blue')
# title
plt.title('ROC curve')
# x label
plt.xlabel('False Positive Rate')
# y label
plt.ylabel('True Positive rate')

plt.legend(loc='best')
plt.savefig('ROC',dpi=300)
plt.show();
```
